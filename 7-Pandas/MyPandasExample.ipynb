# ayengec pandas example with csv file

```python
import pandas as pd
```

We create a dictionary with headers and datas

```python
fpgas  = {"Boards"      :["Basys-2", "Genesys Virtex®-5", "ZedBoard", "DE-10", "Basys-3"], 
          "FPGA Model"  :["Spartan-3E", "Virtex-5", "Zynq-7000", "Cyclone V", "Artix-7"],
          "Brand"       :["Xilinx", "Xilinx", "Xilinx", "Altera", "Xilinx"],
          "Manifacturer":["Digilent", "Digilent", "Digilent", "Terasic", "Digilent"]}
```

```python
myList4csv = pd.DataFrame(fpgas)               # it converts our dictionary to pandas dataframe format
```

```python
myList4csv.to_csv("FPGA_MODELS.csv")             # it creates the .csv file as given name. we can use to_excel method as well. But in my opinion, 
                                                 # if you need to work on excel document, xlrd module is better than pandas
```

```python
rdFile = pd.read_csv("FPGA_MODELS.csv")          # after we saw the FPGA_MODELS.csv file, we can read back all datas with pandas
```
